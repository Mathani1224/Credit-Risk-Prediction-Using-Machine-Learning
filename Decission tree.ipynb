```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
data=pd.read_csv('credit.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
for i in data.columns:
    if(data[i].dtype=='object'):
        data[i]=pd.Categorical(data[i])
data.info()
```

```python
data.describe()
```

```python
replacestruct={"checking_balance":{"< 0 DM":1,"1 - 200 DM":2,"> 200 DM":3,"unknown":-1},
               "credit_history":{"critical":1,"poor":2,"good":3,"very good":4,"perfect":5},
               "savings_balance":{"< 100 DM":1,"100 - 500 DM":2,"500 - 1000 DM":3,"> 1000 DM":4,"unknown":-1},
              "employment_duration":{"unemployed":1,"< 1 year":2,"1 - 4 years":3,"4 - 7 years":4,"> 7 years":5},
              "phone":{"yes":2,"no":1},
              "default":{"no":0,"yes":1}}
```

```python
c=["purpose","housing","other_credit","job"]
```

```python
data=data.replace(replacestruct)
```

```python
data=pd.get_dummies(data,columns=c)
```

```python
data.head(20)
```

```python
x=data.drop("default",axis=1)
y=data['default']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
model=DecisionTreeClassifier(criterion='gini',random_state=1)
```

```python
model.fit(x_train,y_train)
```

```python
print(model.score(x_train,y_train))
print(model.score(x_test,y_test))
```

```python
dtree=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=1)
```

```python
dtree.fit(x_train,y_train)
```

```python
print(dtree.score(x_train,y_train))
print(dtree.score(x_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_random=RandomForestClassifier(n_estimators=50,random_state=1)
```

```python
model_random.fit(x_train,y_train)
```

```python
print(model_random.score(x_test,y_test))
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
model_bagging=BaggingClassifier(estimator=dtree,n_estimators=50,random_state=1)
```

```python
model_bagging.fit(x_train,y_train)
```
